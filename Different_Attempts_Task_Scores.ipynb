```python
import matplotlib.pyplot as plt

# Manish's scores
scores = [3, 3.5, 5]

# Attempts
attempts = ['First Attempt', 'Second Attempt', 'Third Attempt']

# Plotting the bar chart
plt.bar(attempts, scores, color=['red', 'blue', 'green'],edgecolor='black')
plt.plot(attempts, scores, marker='o', color='purple', linestyle='-', linewidth=2, markersize=2)
plt.scatter(attempts,scores,edgecolor='orange')
plt.xlabel('Attempts')

plt.ylabel('Scores')
plt.title('Manish\'s Scores in Different Attempts')
plt.ylim(0, 5.5)  
plt.show()
```

```python
import matplotlib.pyplot as plt

# Manish's scores
Task_scores = [3, 3.5, 5]

# Attempts
attempts = ['First Attempt', 'Second Attempt', 'Third Attempt']

# Plotting the line chart
plt.plot(attempts, Task_scores, marker='o', color='b', linestyle='-', linewidth=2, markersize=8)
plt.xlabel('Attempts',color='g',size=15)
plt.ylabel('Scores',color='g',size=15)
plt.title('Manish\'s Scores in Different Attempts',size=15,color='g')
plt.ylim(0, 5.5)  
plt.show()
```
